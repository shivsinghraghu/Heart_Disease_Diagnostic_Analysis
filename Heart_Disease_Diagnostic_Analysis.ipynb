```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "/content/drive/MyDrive/Heart Disease data.csv"
data = pd.read_csv(url)
```

```python
# Plot Heart disease rates
plt.figure(figsize=(10, 6))
sns.countplot(x='target', data=data, palette='pastel')
plt.title('Heart Disease Rates')
plt.xlabel('Target (0 = No Heart Disease, 1 = Heart Disease)')
plt.ylabel('Count')
plt.show()
```

```python
# Plot Heart disease by gender
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', hue='target', data=data, palette='pastel')
plt.title('Heart Disease by Gender')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.legend(title='Target', labels=['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
# Plot Heart disease by age
plt.figure(figsize=(12, 6))
sns.histplot(x='age', hue='target', data=data, palette='pastel', kde=True)
plt.title('Heart Disease by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Target', labels=['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
# Compare attributes to extract necessary information
plt.figure(figsize=(10, 6))
sns.scatterplot(x='chol', y='trestbps', hue='target', data=data, palette='pastel')
plt.title('Cholesterol vs Resting Blood Pressure')
plt.xlabel('Cholesterol')
plt.ylabel('Resting Blood Pressure')
plt.legend(title='Target', labels=['No Heart Disease', 'Heart Disease'])
plt.show()
```
